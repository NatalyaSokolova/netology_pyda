## Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
df=pd.read_csv(r'C:\Users\Natasha\Downloads\Python_13_join\ml-latest-small\ratings.csv')
df.head()
```

```python
#Делаем фильтр по количеству выставленных рейтингов, в разрезе userId
df1=df.groupby('userId').count()
z=df1[df1.rating > 100] 
print(z)
len(z)
```

```python
#Определяем список id пользователей, которые поставили >100 рейтингов
list_of_frequent_users=z.filter(items = ['userId'])
list_of_frequent_users
```

```python
#делаем таблицу с минмумами и максимумами по датам
d=df.groupby('userId').agg(['min', 'max'])[['timestamp']]
d.head()
```

```python
#Считаем среднее время "жизни" пользователей (всех) в разрезе id
start_date = df.groupby('userId')[['timestamp']].min()
end_date=df.groupby('userId')[['timestamp']].max()

all_users_lt = start_date.copy(deep=True)
del all_users_lt['timestamp']

all_users_lt['livetime'] = end_date - start_date
all_users_lt
```

```python
#Объединяем 2 таблицы. Пришлось делать через merge, потому что через join получилось слишком мало значений :(
final_table=pd.merge(list_of_frequent_users, all_users_lt, on='userId', how='left')
final_table
```

```python
#считаем среднее время "жизни" нужной группы пользователей
final_table['livetime'].mean()
```
